###Tratamento de dados com PANDAS



```python
import pandas as pandinha
import numpy as numpynho
```

```python
#Importando os dados do CSV

tabela_valores = pandinha.read_csv('https://raw.githubusercontent.com/celsocrivelaro/simple-datasets/main/seeds.csv', header=None, index_col=None)
tabela = pandinha.DataFrame(tabela_valores)
tabela
```

```python
#remover colunas extras, linhas com valores nulos

tabela.drop(columns = [6, 7, 8], inplace = True)

print(len(tabela.columns))
```

```python
#Inserindo as linhas de cabeçalho e excluindo linhas com valores nulos

tabela.columns = ['Área A', 'Perímetro P', 'Extensão do núcleo', 'Largura', 'Coef. de Assimetria', 'Extensão do sulgo do núcleo']
tabela.dropna(inplace=True)
print("Anterior:", len(tabela_valores), "Nova Tabela:", len(tabela))
tabela
```

```python
#Adicionando coluna compactação que receberá um calculo tal que Compactação = 4*pi*AREA/PERIMETRO²

tabela['Compactação'] = (tabela['Área A'] * 4 * numpynho.pi)/(tabela['Perímetro P']**2)
tabela
```

```python
#Exportando a tabela para CSV

from google.colab import files

tabela.to_csv('tabela_final.csv', encoding = 'utf-8-sig')
files.download('tabela_final.csv')
```
